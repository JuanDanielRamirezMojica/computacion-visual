## Importar las libreras necesarias

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

#Para el bonus con sliders
import ipywidgets as widgets
from IPython.display import display
```

## Subir la imagen

```python
# Subir una imagen desde el pc
uploaded = files.upload()

#Obtener nombre del archivo
image_path = list(uploaded.keys())[0]

# Cargar la imagen
original = cv2.imread(image_path)
```

## Mostrar la imagen original

```python
# Funcin para mostrar imagen y descargar

def show_download_img(image, name, title,):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()

    # Guardar la imagen procesada
    output_path = '/content/'+name+'.jpg'
    cv2.imwrite(output_path, image)

    # Descargar la imagen procesada
    files.download(output_path)
```

```python
show_download_img(original,'original' ,'Imagen Original')
```

# Mostrar los canales RGB y HSV por separado.

```python
## Separar canales
r, g, b = cv2.split(original)
hsv = cv2.cvtColor(original, cv2.COLOR_RGB2HSV)
h, s, v = cv2.split(hsv)
```

**NOTA:** Cuando se separan los canales R, G y B de una imagen, cada canal es una matriz en escala de grises (sin color real). Por eso, al visualizarlo, es necesario usar un parmetro como cmap='Reds' para representar el canal rojo de forma visual. Esto es solo un truco de coloracin para mejor interpretacin en matplotlib. Si se desea ver el canal como un color real (por ejemplo, rojo puro), se debe crear una nueva imagen RGB donde solo se active el canal correspondiente.

# Utilizar slicing de matrices para modificar regiones especficas de la imagen

```python
plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')
plt.axis('off')
plt.show()
```

```python
# Copiar imagen para modificar
imagen_modificada = cv2.cvtColor(original, cv2.COLOR_BGR2RGB).copy()

# Cambiar el color de varias reas rectangulares
# Primer cuadro (rojo)
imagen_modificada[50:150, 50:150] = [255, 0, 0]  # Rojo puro

# Segundo cuadro (verde)
imagen_modificada[200:300, 100:200] = [0, 255, 0]  # Verde puro

# Tercer cuadro (azul)
imagen_modificada[100:200, 300:400] = [0, 0, 255]  # Azul puro

# Cuarto cuadro (amarillo)
imagen_modificada[250:350, 250:350] = [255, 255, 0]  # Amarillo


# Tercer cuadro (azul)
imagen_modificada[100:200, 500:900] = [0, 255, 255]  # Cian

# Cuarto cuadro (amarillo)
imagen_modificada[700:850, 850:950] = [255, 255, 127]  # Amarillo


# Mostrar cambio de regiones
show_download_img(imagen_modificada,'mod_colorblocks', 'Imagen con Mltiples Regiones Modificadas')
# plt.imshow(imagen_modificada)
# plt.title('Imagen con Mltiples Regiones Modificadas')
# plt.axis('off')
# plt.show()
```

# Sustituir

```python
# Copiar una regin (Ejemplo: de (50:150, 50:150)) a otra parte

original = cv2.cvtColor(original, cv2.COLOR_BGR2RGB).copy()
region = original[700:900, 700:900].copy()
imagen_modificada[700:900, 200:400] = region

# Mostrar imagen despus de copiar regin
show_download_img(imagen_modificada,'copybabyface', 'Imagen con Regin Copiada')


# plt.imshow(imagen_modificada)
# plt.title('Imagen con Regin Copiada')
# plt.axis('off')
# plt.show()
```

```python

# Copiar imagen original
original = cv2.cvtColor(original, cv2.COLOR_BGR2RGB).copy()
slicingsustitucion = original.copy()

# Dividir la imagen en dos mitades
height, width, _ = original.shape
left_half = original[:, :width // 2]  # Tomamos la mitad izquierda completa

# Sustituir una regin: mitad derecha = mitad izquierda
slicingsustitucion[:, width // 2:] = left_half

# Mostrar la imagen modificada
show_download_img(slicingsustitucion,'halflefthalfright', 'Mitad Izquierda Sustituye la Derecha')
# plt.imshow(cv2.cvtColor(slicingsustitucion, cv2.COLOR_BGR2RGB))
# plt.title('Mitad Izquierda Sustituye la Derecha')
# plt.axis('off')
# plt.show()
```

```python
# Histograma de la imagen en escala de grises
imagen_gris = cv2.cvtColor(original, cv2.COLOR_RGB2GRAY)

plt.figure(figsize=(8,6))
plt.hist(imagen_gris.ravel(), bins=256, range=[0,256], color='gray')
plt.title('Histograma de Intensidades (Escala de Grises)')
plt.xlabel('Intensidad de pxel')
plt.ylabel('Cantidad de pxeles')
plt.grid(True)
plt.show()

plt.savefig('histograma_grises.png')
files.download('histograma_grises.png')
```

```python

# Funcin manual
def ajustar_brillo_contraste(img, alpha=10.7, beta=7):
    """ Ajusta el brillo y contraste de una imagen.
    alpha: factor de contraste
    beta: factor de brillo
    """
    nueva_img = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
    return nueva_img

original = cv2.cvtColor(original, cv2.COLOR_BGR2RGB).copy()

# Aplicar ajuste
imagen_ajustada = ajustar_brillo_contraste(original, alpha=2.5, beta=50)

# Mostrar imagen ajustada
show_download_img(imagen_ajustada,'brilloContraste', 'Imagen con Brillo y Contraste Ajustados')


# plt.imshow(imagen_ajustada)
# plt.title('Imagen con Brillo y Contraste Ajustados')
# plt.axis('off')
# plt.show()
```

```python
# Bonus: Sliders para brillo y contraste
original = cv2.cvtColor(original, cv2.COLOR_BGR2RGB).copy()



# Funcin para modificar brillo y contraste
def brightness_contrast(brightness=0, contrast=1.0):

    img = original.copy()

    # Aplicar ajustes
    img = cv2.convertScaleAbs(img, alpha=contrast, beta=brightness)

    # Mostrar imagen resultante
    plt.figure(figsize=(6,6))
    plt.imshow(img)
    plt.title(f'Brillo: {brightness}, Contraste: {contrast:.2f}')
    plt.axis('off')
    plt.show()

# Crear sliders
brightness_slider = widgets.IntSlider(
    value=0,
    min=-100,
    max=100,
    step=1,
    description='Brillo:',
    continuous_update=False
)


contrast_slider = widgets.FloatSlider(
    value=1.0,
    min=0.1,
    max=3.0,
    step=0.1,
    description='Contraste:',
    continuous_update=False
)

#Ac se configura la UI
ui = widgets.VBox([brightness_slider, contrast_slider])


# Output interactivo
out = widgets.interactive_output(
    brightness_contrast,
    {'brightness': brightness_slider, 'contrast': contrast_slider}
)

# Mostrar interfaz
display(ui, out)
```
