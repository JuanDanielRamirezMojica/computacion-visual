## Importar las libreras necesarias

```python
from google.colab import files #Subir imagenes
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Subir la imagen

```python
# Subir una imagen desde el pc
uploaded = files.upload()

#Obtener nombre del archivo
image_path = list(uploaded.keys())[0]

# Cargar la imagen
image = cv2.imread(image_path)
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.title("Imagen original")
plt.axis('off')
plt.show()
```

## Cambiar de color la imagen

El hue (tono) es el valor que representa el color puro en un modelo de color como HSV.
Es decir, define si el color es rojo, azul, verde, amarillo, etc.
Sigue este patrn:

0  rojo

30  amarillo

60  verde

90  cian

120  azul

150  rosa/magenta

```python
# Cambiar el color de la imagen usando HSV
hsv_img = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)  # Convertir la imagen de RGB a HSV

# Definir un nuevo hue
hue_shift = 30
hsv_img[:, :, 0] = hue_shift  # Cambiar el canal de tono (Hue)

# Convertir de regreso a RGB para visualizar
newcolor_image = cv2.cvtColor(hsv_img, cv2.COLOR_HSV2RGB)

# Mostrar la imagen
plt.imshow(newcolor_image)
plt.title("Imagen con Cambio de Color Amarillo")
plt.axis('off')
plt.show()


# Guardar la imagen procesada
output_path = '/content/yellow_image.jpg'
cv2.imwrite(output_path, newcolor_image)


# Descargar la imagen procesada
files.download(output_path)
```

## Convertir la imagen a escala de grises

```python

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Guardar la imagen procesada (en escala de grises)
output_path = '/content/gray_image.jpg'
cv2.imwrite(output_path, gray_image)

# Visualizar la imagen en escala de grises
plt.imshow(gray_image, cmap='gray')
plt.title("Imagen en Escala de Grises")
plt.axis('off')  # Desactiva los ejes
plt.show()

# Descargar la imagen procesada
files.download(output_path)
```

## Aplicar filtro de desenfoque (blur)

```python
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)


# Guardar la imagen procesada (en blurred)
output_path = '/content/blurred_image.jpg'
cv2.imwrite(output_path, gray_image)

# Mostrar la imagen con desenfoque
plt.imshow(blurred_image, cmap='gray')
plt.title("Imagen con Filtro de Desenfoque")
plt.axis('off')
plt.show()

# Descargar la imagen procesada
files.download(output_path)
```

## Filtro de afilado (sharpening) - para realzar los bordes


```python
kernel_sharpen = np.array([[0, -1, 0], [-1, 5,-1], [0, -1, 0]])
sharpened_image = cv2.filter2D(gray_image, -1, kernel_sharpen)


# Guardar la imagen procesada (afialada)
output_path = '/content/afilada_image.jpg'
cv2.imwrite(output_path, sharpened_image)


# Mostrar la imagen afilada
plt.imshow(sharpened_image, cmap='gray')
plt.title("Imagen con Filtro de Afilado")
plt.axis('off')
plt.show()

# Descargar la imagen procesada
files.download(output_path)
```

##  Deteccin de bordes usando el filtro de Sobel

```python
# Filtro Sobel en X
sobel_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)
sobel_x = cv2.convertScaleAbs(sobel_x)

# Filtro Sobel en Y
sobel_y = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)
sobel_y = cv2.convertScaleAbs(sobel_y)

# Combinar los bordes de Sobel en X y Y
sobel_edges = cv2.addWeighted(sobel_x, 0.5, sobel_y, 0.5, 0)


# Guardar la imagen procesada (bordes)
output_path = '/content/bordes_sobel_image.jpg'
cv2.imwrite(output_path, sobel_edges)


# Mostrar los bordes detectados con Sobel
plt.imshow(sobel_edges, cmap='gray')
plt.title("Deteccin de Bordes con Sobel")
plt.axis('off')
plt.show()


# Descargar la imagen procesada
files.download(output_path)
```

## Deteccin de bordes con el filtro Laplaciano

```python
laplacian_edges = cv2.Laplacian(gray_image, cv2.CV_64F)
laplacian_edges = cv2.convertScaleAbs(laplacian_edges)


# Guardar la imagen procesada (bordes laplace)
output_path = '/content/bordes_laplace_image.jpg'
cv2.imwrite(output_path, laplacian_edges)


# Mostrar los bordes detectados con Laplaciano
plt.imshow(laplacian_edges, cmap='gray')
plt.title("Deteccin de Bordes con Laplaciano")
plt.axis('off')
plt.show()


# Descargar la imagen procesada
files.download(output_path)
```

## Comparar resultados

```python
#(Sobel vs Laplaciano)
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(sobel_edges, cmap='gray')
axes[0].set_title("Sobel Edges")
axes[0].axis('off')

axes[1].imshow(laplacian_edges, cmap='gray')
axes[1].set_title("Laplacian Edges")
axes[1].axis('off')

plt.show()
```

## Agregar sliders (cv2.createTrackbar) para modificar en vivo parmetros como tamao del kernel o tipo de filtro aplicado.

Nota: Dado que cv2.createTrackbar no funciona directamente en Colab, porque depende de ventanas OpenCV (cv2.imshow) que requieren GUI, se deben usar widgets de ipywidgets para realizar este punto.

```python
# Bonus: Selector de imagen + Sliders dinmicos para filtros
import ipywidgets as widgets
from IPython.display import display, clear_output



# Funcin para aplicar filtros segn los parmetros
def process_image(filter_type, kernel_size):
    plt.figure(figsize=(6,6))

    if filter_type == 'Original':
        # No influye el slider.
        img = image
        plt.title("Imagen Original - Slider NO necesario")


    elif filter_type == 'Grises':
        # No influye el slider.
        img = gray_image
        plt.title("Imagen en Escala de Grises  - Slider NO necesario")


    elif filter_type == 'Desenfocada':
        # Asegurarse que kernel_size sea impar y mayor a 0, debe ser impar para obtener un centro.
        k = kernel_size if kernel_size % 2 == 1 else kernel_size + 1
        blurred = cv2.GaussianBlur(gray_image, (k, k), 0)
        img = blurred
        plt.title(f"Desenfoque (Kernel {k}x{k})")




    elif filter_type == 'Afilada':
        # No influye el slider.
        sharpened = cv2.filter2D(gray_image, -1, np.array([[0, -1, 0], [-1, 5,-1], [0, -1, 0]]))
        img = sharpened
        plt.title("Imagen Afilada - Slider NO necesario")


    elif filter_type == 'Bordes Sobel':
        sobel_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=kernel_size)
        sobel_y = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=kernel_size)
        sobel_edges = cv2.addWeighted(cv2.convertScaleAbs(sobel_x), 0.5, cv2.convertScaleAbs(sobel_y), 0.5, 0)
        img = sobel_edges
        plt.title(f"Bordes Sobel (Kernel {kernel_size}x{kernel_size})")



    elif filter_type == 'Bordes Laplaciano':
        laplacian_edges = cv2.Laplacian(gray_image, cv2.CV_64F, ksize=kernel_size)
        img = cv2.convertScaleAbs(laplacian_edges)
        plt.title(f"Bordes Laplaciano (Kernel {kernel_size}x{kernel_size})")

    # Mostrar imagen procesada
    if len(img.shape) == 2:  # Escala de grises
        plt.imshow(img, cmap='gray')
    else:  # Imagen a color
        plt.imshow(img)

    plt.axis('off')
    plt.show()

# Crear widgets
filter_selector = widgets.Dropdown(
    options=['Original', 'Grises', 'Desenfocada', 'Afilada', 'Bordes Sobel', 'Bordes Laplaciano'],
    value='Original',
    description='Filtro:',
    style={'description_width': 'initial'}
)

kernel_slider = widgets.IntSlider(
    value=3,
    min=1,
    max=15,
    step=2,
    description='Kernel Size:',
    continuous_update=False
)

# Funcin interactiva
ui = widgets.VBox([filter_selector, kernel_slider])

out = widgets.interactive_output(process_image, {'filter_type': filter_selector, 'kernel_size': kernel_slider})

# Mostrar UI
display(ui, out)
```
